### Q1. Perform the following computation without use of any libraries


```math
    [1,-40,36]
A = [2,80,-72]
    [3,-40,-72]

    [0,49,7]
B = [9,64,81]
    [222,3,0.1]
```

Given two matrices A & B perform the operations:
* A + B
* A - B
* AB

````python
A = [[1,-40,36],[2,80,-72],[3,-40,-72]]
B = [[0,49,7],[9,64,81],[222,3,0.1]]
A_add_B = []

for i in range(len(A)):
    temp = []
    for j in range(len(B)):
        temp.append(A[i][j] + B[i][j])
    A_add_B.append(temp)
        
print("A+B = ", A_add_B)

A_sub_B = []

for i in range(len(A)):
    temp = []
    for j in range(len(B)):
        temp.append(A[i][j] - B[i][j])
    A_sub_B.append(temp)
        
print("A-B = ", A_sub_B)

A_mul_B = []
for i in range(len(A)):
    temp = []
    temp_sum = 0
    for j in range(len(A[0])):
        for k in range(len(B)):
            temp_sum = temp_sum + (A[i][k] * B[k][j])
        temp.append(temp_sum)
        temp_sum = 0
    A_mul_B.append(temp)
        
print("AB = ", A_mul_B)
````

### Q2. Perform the same computation using Numpy package

````python
import numpy 

A = numpy.array([[1,-40,36],[2,80,-72],[3,-40,-72]])
B = numpy.array([[0,49,7],[9,64,81],[222,3,0.1]])

A_add_B = numpy.add(A,B)
print("A+B = ", A_add_B)

A_sub_B = numpy.subtract(A,B)
print("A-B = ", A_sub_B)

A_mul_B = numpy.dot(A,B)
print("AB = ", A_mul_B)
````

### Q3. Compute using numpy
* A<sup>T</sup>
* B<sup>T</sup>
* A<sup>T</sup>B<sup>T</sup>
* AA<sup>T</sup>

````python
print("A transpose = ", A.T)
print("B transpose = ", B.transpose())

A_trans = numpy.transpose(A)
B_trans = numpy.transpose(B)
print("ATBT = ", numpy.dot(A_trans,B_trans))
print("AAT = ", numpy.dot(A,A_trans))
````

### Q10. Use Seaborn "TIPS" dataset and plot pairplot, jointplot, catplot, displot, and HeatMap

"Comments on the different plots you used and why you used them what observations you made. Regarding the Heatmap, comment on observations you can make with a heat map and do you see any correlation in the data in consideration, which normalisation technique you used."

````python
#from google.colab import drive
#drive.mount('/content/drive')
````

````python
tips = sns.load_dataset("tips")
tips.head()
````

````python
plt.figure(figsize=(8,8))
sns.histplot(data=tips['day']) 
# histogram is a simple bar graph like plotting 
# here it is plotted to display the number of tips given on days of a week
# we observe that most of the tips are given on Saturdays and lowest on Fridays
````

````python
sns.histplot(x='day',y='total_bill',data=tips)
````

````python
sns.pairplot(tips,hue='sex',diag_kind='hist') 
# hue is specified to plot that aspect in different color
# pairplot displays pairwise relationship 
# each numeric value in dataset is represented across x and y axes
# following plottings help us understand relation between tip and total_bill, size and total_bill, tip and size.
````

````python
sns.jointplot(x='tip',y='total_bill',data=tips)
# joint plot is combination of univariate an bivariate graphs
# here we get to know the relation between total_bill and tip (bivariate) 
# also we get to know most given tip amount (univariate) 
````

````python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='sex')
# adding hue parameter will give differentiation among the mentioned attribute(sex)
````

````python
sns.catplot(x = 'size', y = 'total_bill', data = tips)
# catplot gives the plotting in categories, even if the values are continuous(total_bill)
# catplot will show them categorically on axes
````

````python
sns.displot(x='size', y='total_bill',data=tips)  
# displot helps in visualizing univariate or bivariate distribution of data 
# following example showa bivariate graph which shows how total_bill is distributed as per size of group
# we observe that most of the bill amount is between 10 to 20 if group size is 2 
# for size 1 none of the bill amount is greater than 12   
````

````python
tips_hm = tips.corr()
print(tips_hm)
plt.figure(figsize=(8,8))
sns.heatmap(tips_hm, cmap="Blues")
# using heatmaps we get to know how the attributes are correlated 
# here we can see total_bill and tip are slightly more correlated as compared tip and size 
````

````python
plt.figure(figsize=(8,8))
tips_hm1 = tips.pivot_table(index='day', columns='sex', values='total_bill')
sns.heatmap(tips_hm1, cmap='Blues')
# using this plot we get to know on what day who has greater total_bill
# on sundays male visitors have highest bill while on fridays female visitors have lease bill amount
````

````python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
LabelEncoder_X= LabelEncoder()
tips['sex']= LabelEncoder_X.fit_transform(tips['sex'])
tips['day']= LabelEncoder_X.fit_transform(tips['day'])
tips['time']= LabelEncoder_X.fit_transform(tips['time'])
tips['smoker']= LabelEncoder_X.fit_transform(tips['smoker'])
````

````python
tips.head()
````

````python
# Normalization 
# used min-max normalization as the output will be [0,1] as required by heatmaps
# normalizing all encoded values and total_bill and tip

from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
norm_values = scaling.fit_transform(tips[['total_bill', 'tip','sex','smoker','day','time']])

print(norm_values)
tips['total_bill'] = norm_values[:,0]
tips['tip'] = norm_values[:,1]
tips['sex'] = norm_values[:,2]
tips['smoker'] = norm_values[:,3]
tips['day'] = norm_values[:,4]
tips['time'] = norm_values[:,5]

tips_hm = tips.corr()
plt.figure(figsize=(8,8))
sns.heatmap(tips_hm, cmap="Blues")
````

````python
# from the above plot we can conclude that size and total_bill are highly correlated  
# tip and size are second highly correlated
````
